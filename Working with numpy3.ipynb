```python
#stacking the data
arr1 = np.array([1,2,3,4])
arr2 = np.array([10,20,30,40])


arr3 =  np.vstack((arr1,arr2))
print (f"The shape of arr3 is {arr3.shape}")



arr4 = np.hstack((arr1,arr2))
print(f"The shape of arr4 is {arr4.shape}")
print(arr4)
```
